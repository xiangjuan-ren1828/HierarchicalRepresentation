### NC-reviews: influence of the primed order (on 64 directional transitional pairs

```r
# ------ Only comparison between HubToLeaf and LeafToHub group ------
# clusterResult_pd_bin_pathGroup <- clusterResult_pd_bin[
#   clusterResult_pd_bin$networkName == "0hubToleaf" |
#   clusterResult_pd_bin$networkName == "2leafToHub",
# ]
clusterResult_pd_bin_pathGroup <- clusterResult_pd_bin
clusterResult_pd_bin_pathGroup
```

```r
# ------ Separating the 64 transition pairs into 4 pars according to their exposure order in Learning I for LeafToHub group ------

# Enter the numbers as a matrix (row-wise)
pairSorted <- matrix(
  c(11,10,
    11,12,
    2,6,
    9,8,
    14,8,
    2,5,
    3,5,
    4,5,
    3,7,
    4,7,
    9,7,
    14,7,
    15,16,
    16,15,
    15,10,
    15,13,
    16,6,
    16,7,
    10,11,
    12,11,
    10,15,
    1,10,
    1,12,
    10,1,
    12,1,
    1,6,
    12,13,
    1,5,
    10,7,
    12,7,
    6,2,
    8,9,
    8,14,
    6,16,
    13,15,
    6,1,
    13,12,
    8,13,
    13,8,
    6,5,
    8,5,
    13,5,
    6,7,
    8,7,
    13,7,
    5,2,
    5,3,
    5,4,
    5,1,
    5,6,
    5,8,
    5,13,
    5,7,
    7,3,
    7,4,
    7,9,
    7,14,
    7,16,
    7,10,
    7,12,
    7,6,
    7,8,
    7,13,
    7,5),
  byrow = TRUE, ncol = 2
)

# Turn into a data frame with names
pairs_df <- as.data.frame(pairSorted)
colnames(pairs_df) <- c("fromNode", "toNode")
# Add an order column (handy if you want to restore the input order later)
pairs_df$order <- seq_len(nrow(pairs_df))

pairs_df
```

```r
# Split indices into 4 parts
idx_list <- split(seq_len(nrow(pairs_df)),
                  cut(seq_len(nrow(pairs_df)), 4, labels = FALSE))

# Prepare an empty list to store results
res_list <- vector("list", length = 4)

# Loop over the 4 parts
for (i in 1:4) {
  part_df <- pairs_df[idx_list[[i]], , drop = FALSE]
  
  # Extract matching rows from clusterResult_pd_bin
  sel <- paste(clusterResult_pd_bin_pathGroup$targetObj, clusterResult_pd_bin_pathGroup$respObj0) %in%
         paste(part_df$fromNode, part_df$toNode)
    
  res_list[[i]] <- clusterResult_pd_bin_pathGroup[sel, , drop = FALSE]
}

# You can now access each result:
res_part1 <- res_list[[1]]
res_part2 <- res_list[[2]]
res_part3 <- res_list[[3]]
res_part4 <- res_list[[4]]
```

```r
choiceModel <- mixed(choice ~ networkName*trialCnt + (1 + trialCnt | subNo), data = res_part4, family = binomial, method="LRT")
# choiceModel <- mixed(choice ~ networkName + (1 | subNo), data = clusterResult_pd_bin_HubRelated, family = binomial, method="LRT")
choiceModel
```

## post-hoc analysis

### If no interactions

```r
emm1 = emmeans(choiceModel, specs = ~ networkName) 
emm1
```

### If interactions defined in the GLMM

```r
emm1 = emmeans(choiceModel, specs = ~ networkName*trialCnt) # same as: emmeans(choiceModel, ~ networkName) & emmeans(choiceModel, c("networkName"))
emm1
```

```r
emm_pairs <- pairs(emm1, simple = "networkName", adjust = "bonferroni") # adjust = "bonferroni", "tukey"
emm_pairs
```
