## Overall effects: RTs or choices

### GLMM with choice as the dependent variable

mixed (used for the manuscript)
tutorial: https://cran.r-project.org/web/packages/afex/vignettes/afex_analysing_accuracy_data.html#analysis-of-two-studies-with-glmm

followed what I did for the network+path experiments

```r
# column_name = {0: 'groupNo', 1: 'subjNo', 2: 'trialNo', 3: 'dtrNum', 
#                4: 'cue_dtr1_distance', 5: 'cue_dtr2_distance', 6: 'cue_dtr3_distance', 
#                7: 'tgt_dtr1_distance', 8: 'tgt_dtr2_distance', 9: 'tgt_dtr3_distance',
#                10: 'cue_recencyL', 11: 'cue_recencySL', 12: 'cue_recencyTL', 
#                13: 'walkType', 14: 'transType', 15: 'lureIn', 16: 'errorResp', 
#                17: 'cueBdNode', 18: 'tgtBdNode', 19: 'dtr1_BdNode', 20: 'dtr2_BdNode', 21: 'dtr3_BdNode',
#                22: 'RTs', 23: 'choice_oneHot'} 
```

```r
choiceModel <- mixed(choice_oneHot ~ walkType*transType + (1 + walkType + transType | subjNo), data = resp_glmm_pd_iExp, family = binomial, method="LRT")
choiceModel
```

## Influence of lure stimulus for the boundary node

```r
resp_glmm_pd_iExp_lure <- resp_glmm_pd_iExp[
    resp_glmm_pd_iExp$cueBdNode == "1BdNode", 
]
```

```r
# average across the cue_dtr_distance and tgt_dtr_distance
resp_glmm_pd_iExp_lure$cue_dtr_mean_distance <- rowMeans(
  resp_glmm_pd_iExp_lure[, c("cue_dtr1_distance",
                             "cue_dtr2_distance",
                             "cue_dtr3_distance")],
  na.rm = TRUE
)


resp_glmm_pd_iExp_lure$tgt_dtr_mean_distance <- rowMeans(
  resp_glmm_pd_iExp_lure[, c("tgt_dtr1_distance",
                             "tgt_dtr2_distance",
                             "tgt_dtr3_distance")],
  na.rm = TRUE
)
```

```r
choiceModel <- mixed(choice_oneHot ~ lureIn*transType*dtrNum*cue_dtr_mean_distance + (1 + lureIn + transType + dtrNum + cue_dtr_mean_distance || subjNo), data = resp_glmm_pd_iExp_lure, family = binomial, method="LRT")
choiceModel
```

```r
emm1 = emmeans(choiceModel, specs = ~ networkName*trialCnt) # same as: emmeans(choiceModel, ~ networkName) & emmeans(choiceModel, c("networkName"))
emm1
```

```r
emm_pairs <- pairs(emm1, simple = "networkName", adjust = "bonferroni") # adjust = "bonferroni", "tukey"
emm_pairs
```
