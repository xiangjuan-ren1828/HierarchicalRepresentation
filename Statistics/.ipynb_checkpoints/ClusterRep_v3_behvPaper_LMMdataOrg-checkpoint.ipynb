# Data for each experiment (Exp1: hint+sequential display; Exp2: no hint+sequential display; Exp3: hint+shuffled) was saved separately

<div class="alert alert-success">
<b>(Columns) variables</b>:
    
{0: 'groupNo', 1: 'subjNo', 2: 'trialNo' (log), 3: 'dtrNum',  
 4,5,6: 'cue_dtr_distance', 7,8,9: 'tgt_dtr_distance', 10,11,12: 'cue_recency', 
 13: 'walk_type' (random or hamiltonian), 14: 'trans_type' (within or between), 15: 'lureIn', 16: 'errorResp', 
 17,18,19,20,21: 'boundary_nodes' (cue, target and 3 distractors max),
 22: 'RTs', 23: 'choice_oneHot'}

```python
iExp = 2
ExpWord = expList[iExp]
if ExpWord == 'ImplicitExp':
    bhvDataDir = os.path.join(folder, 'FormalExp-LynnNetwork-Results')
elif ExpWord == 'ExplicitExp':
    bhvDataDir = os.path.join(folder, 'FormalExp-ExplicitLearning-LynnNetwork-Results')
elif ExpWord == 'ImplicitRandExp':
    bhvDataDir = os.path.join(folder, 'FormalExp-LynnNetwork-ImplicitRandom-Results')
    
```

```python
respMat = loadmat(os.path.join(bhvDataDir, f"{ExpWord}_glmm_subj.mat")) # 'glmm_subj'
resp_glmm_subj = respMat['glmm_subj']
resp_glmm_pd    = pd.DataFrame(resp_glmm_subj)
resp_glmm_pd
```

```python
column_name = {0: 'groupNo', 1: 'subjNo', 2: 'trialNo', 3: 'dtrNum', 
               4: 'cue_dtr1_distance', 5: 'cue_dtr2_distance', 6: 'cue_dtr3_distance', 
               7: 'tgt_dtr1_distance', 8: 'tgt_dtr2_distance', 9: 'tgt_dtr3_distance',
               10: 'cue_recencyL', 11: 'cue_recencySL', 12: 'cue_recencyTL', 
               13: 'walkType', 14: 'transType', 15: 'lureIn', 16: 'errorResp', 
               17: 'cueBdNode', 18: 'tgtBdNode', 19: 'dtr1_BdNode', 20: 'dtr2_BdNode', 21: 'dtr3_BdNode',
               22: 'RTs', 23: 'choice_oneHot'} 
resp_glmm_pd = resp_glmm_pd.rename(columns = column_name) 
resp_glmm_pd.groupNo     = resp_glmm_pd.groupNo.astype('int64')
resp_glmm_pd.subjNo      = resp_glmm_pd.subjNo.astype('int64')
resp_glmm_pd.dtrNum      = resp_glmm_pd.dtrNum.astype('int64')
resp_glmm_pd.walkType    = resp_glmm_pd.walkType.astype('int64')
resp_glmm_pd.transType   = resp_glmm_pd.transType.astype('int64')
resp_glmm_pd.lureIn      = resp_glmm_pd.lureIn.astype('int64')
resp_glmm_pd.errorResp   = resp_glmm_pd.errorResp.astype('int64')
resp_glmm_pd.cueBdNode   = resp_glmm_pd.cueBdNode.astype('int64')
resp_glmm_pd.tgtBdNode   = resp_glmm_pd.tgtBdNode.astype('int64')
resp_glmm_pd.dtr1_BdNode = resp_glmm_pd.dtr1_BdNode.astype('int64')
resp_glmm_pd.dtr2_BdNode = resp_glmm_pd.dtr2_BdNode.astype('int64')
resp_glmm_pd.dtr3_BdNode = resp_glmm_pd.dtr3_BdNode.astype('int64')
resp_glmm_pd
```

```python
# column_name = {0: 'groupNo', 1: 'subjNo', 2: 'trialNo', 3: 'dtrNum', 
#                4: 'cue_dtr1_distance', 5: 'cue_dtr2_distance', 6: 'cue_dtr3_distance', 
#                7: 'tgt_dtr1_distance', 8: 'tgt_dtr2_distance', 9: 'tgt_dtr3_distance',
#                10: 'cue_recencyL', 11: 'cue_recencySL', 12: 'cue_recencyTL', 
#                13: 'walkType', 14: 'transType', 15: 'lureIn', 16: 'errorResp', 
#                17: 'cueBdNode', 18: 'tgtBdNode', 19: 'dtr1_BdNode', 20: 'dtr2_BdNode', 21: 'dtr3_BdNode',
#                22: 'RTs', 23: 'choice_oneHot'} 
nWalks   = 2 # Random and Hamiltonian walk
nTrans   = 2 # within and between transition
nLures   = 2 # with or without lures
nError   = 2 # correct or incorrect response based on the final click
nBdNodes = 2 # is boundary node or not
walksList  = ['0Random', '1Hamiltonian']
transList  = ['0Within', '1Between']
luresList  = ['0NoLure', '1WithLure']
errorList  = ['0Correct', '1Incorrect']
BdNodeList = ['0NoBdNode', '1BdNode']

for iW in range(0, nWalks):
    resp_glmm_pd.loc[(resp_glmm_pd['walkType'] == iW), 'walkType'] = walksList[iW]

for iT in range(0, nTrans):
    resp_glmm_pd.loc[(resp_glmm_pd['transType'] == iT), 'transType'] = transList[iT]

for iL in range(0, nLures):
    resp_glmm_pd.loc[(resp_glmm_pd['lureIn'] == iL), 'lureIn'] = luresList[iL]

for iE in range(0, nError):
    resp_glmm_pd.loc[(resp_glmm_pd['errorResp'] == iE), 'errorResp'] = errorList[iE]

for iB in range(0, nBdNodes):
    # ------ cueBdNode ------
    resp_glmm_pd.loc[(resp_glmm_pd['cueBdNode'] == iB), 'cueBdNode'] = BdNodeList[iB]
    # ------ tgtBdNode ------
    resp_glmm_pd.loc[(resp_glmm_pd['tgtBdNode'] == iB), 'tgtBdNode'] = BdNodeList[iB]
    # ------ dtr1_BdNode ------
    resp_glmm_pd.loc[(resp_glmm_pd['dtr1_BdNode'] == iB), 'dtr1_BdNode'] = BdNodeList[iB]
    # ------ dtr2_BdNode ------
    resp_glmm_pd.loc[(resp_glmm_pd['dtr2_BdNode'] == iB), 'dtr2_BdNode'] = BdNodeList[iB]
    # ------ dtr3_BdNode ------
    resp_glmm_pd.loc[(resp_glmm_pd['dtr3_BdNode'] == iB), 'dtr3_BdNode'] = BdNodeList[iB]

resp_glmm_pd.loc[(resp_glmm_pd['groupNo'] == iExp + 1), 'groupNo'] = ExpWord

resp_glmm_pd
```

```python
path_or_buf = os.path.join(bhvDataDir, f"{ExpWord}_glmm_subj.csv")
path_or_buf
```

```python
resp_glmm_pd.to_csv(path_or_buf = path_or_buf)
```
